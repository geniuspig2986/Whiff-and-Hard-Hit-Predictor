```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline
```

```python

# Load your dataset
df = pd.read_csv("statcast_pitch_swing_data_20240402_20241030_with_arm_angle.csv")

# Take a quick look
print(df.shape)
df.head()
```

```python
# Select key columns for batters
columns = [
    'game_date', 'player_name', 'events', 'launch_speed', 'launch_angle',
    'hit_distance_sc', 'estimated_ba_using_speedangle', 'estimated_woba_using_speedangle',
    'woba_value', 'woba_denom', 'game_year'
]
df = df[columns]

# Drop missing or null rows
df = df.dropna(subset=['player_name', 'launch_speed', 'events'])

# Convert dates to datetime
df['game_date'] = pd.to_datetime(df['game_date'])
```
